### Создание таблиц

```python
cursor.execute("""create table transactions ( 
    transaction_id int,
    customer_id int,
    product_id int,
    transaction_date timestamp,	
    online_order bool,
    order_status varchar)""")
```

```python
cursor.execute("select * from transactions")
```

```python
conn.commit()
```

```python
cursor.execute("""create table products ( 
    id int,
    brand varchar,	
    product_line varchar,
    product_class varchar,	
    product_size varchar,
    list_price float,	
    standard_cost float)""")
```

```python
conn.commit()
```

```python
cursor.execute("""create table customers ( 
    id int,
    first_name varchar,
    last_name varchar, 
    gender varchar,
    DOB date,
    job_title varchar,
    job_industry_category varchar,
    wealth_segment varchar,
    deceased_indicator varchar,
	owns_car varchar,
    address varchar)""")
```

```python
conn.commit()
```

```python
cursor.execute("""create table addresses ( 
    address varchar,
    postcode int,
    state varchar,
    country varchar,
    property_valuation int)""")
```

```python
conn.commit()
```

```python
cursor.execute("""alter table transactions add primary key(transaction_id)""")
```

```python
conn.commit()
```

```python
cursor.execute("""alter table addresses add primary key(address)""")
cursor.execute("""alter table customers add primary key(id)""")
cursor.execute("""alter table products add primary key(id)""")
```
